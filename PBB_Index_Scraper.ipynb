```python
import requests, pandas as pd
url='https://raw.githubusercontent.com/jamieprince/jamieprince.github.io/main/bloomberg_billionaire_index.html'
from bs4 import BeautifulSoup
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')
```

```python
soup.findAll('div')
table=soup.find_all("div",class_="table-cell t-name")
for i in soup.find_all('div', class_="table-cell t-name"):
    print(i.a)
name_list= [i.text for i in table]
name_list = [item.replace("\r\n                            ", "").replace(",", "") for item in name_list]
name_list
```

```python
table=soup.find_all("div",class_="table-cell active t-nw")
for i in soup.find_all('div', class_="table-cell active t-nw"):
    print(i.a)
networth_list= [i.text for i in table]
networth_list= [item2.replace("\r\n                        ", "").replace("\r\n                    ", "").replace("$","").replace("B","") for item2 in networth_list]
del networth_list[0]
import numpy as np
list(np.float_(networth_list))
networth_list
```

```python
table=soup.find_all("div",class_="table-cell t-country")
for i in soup.find_all('div', class_="table-cell t-country"):
    print(i.a)
country_list= [i.text for i in table]
country_list = [item3.replace("\r\n                        ", "").replace("\r\n                    ", "") for item3 in country_list]
del country_list[0]
country_list
```

```python
table=soup.find_all("div",class_="table-cell t-industry")
for i in soup.find_all('div', class_="table-cell t-industry"):
    print(i.a)
industry_list = [i.text for i in table]
industry_list = [item4.replace("\r\n                        ", "").replace("\r\n                    ", "") for item4 in industry_list]
del industry_list[0]
industry_list
```

```python
df1=pd.DataFrame([name_list,networth_list,country_list,industry_list]).T
```

```python
df1.columns=['Name','Total net worth','Country / Region','Industry']
```

```python
pd.concat([df1])
```

```python
df1.to_csv('Bloomberg_Billionaires_Index.csv')
```
